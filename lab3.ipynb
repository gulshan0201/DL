3 Write a program to implement artificial neural networks without back propagation 

```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1):
        # Initialize weights and bias randomly
        self.weights = np.random.randn(input_size)
        self.bias = np.random.randn()
        self.lr = learning_rate

    def activation(self, x):
        # Step Function: returns 1 if x >= 0, else 0
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        # Calculation: (Inputs * Weights) + Bias
        summation = np.dot(inputs, self.weights) + self.bias
        return self.activation(summation)

    def train(self, training_data, labels, epochs=10):
        print("Starting training...")
        for epoch in range(epochs):
            total_error = 0
            for i in range(len(training_data)):
                # 1. Forward Pass
                prediction = self.predict(training_data[i])
                
                # 2. Calculate Error (Target - Actual)
                error = labels[i] - prediction
                total_error += abs(error)
                
                # 3. Weight Update (Delta Rule - No Backprop needed)
                # New Weight = Old Weight + (Learning Rate * Error * Input)
                self.weights += self.lr * error * training_data[i]
                self.bias += self.lr * error
                
            print(f"Epoch {epoch+1}: Total Errors = {total_error}")
            if total_error == 0:
                print("Model has converged!")
                break

# --- Execution ---

# Logic Gate Data (AND Gate)
# Inputs: [0,0], [0,1], [1,0], [1,1]
# Labels: 0, 0, 0, 1
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
labels = np.array([0, 0, 0, 1])

# Initialize and Train
nn = Perceptron(input_size=2)
nn.train(inputs, labels, epochs=20)

# Test the model
print("\nTesting Predictions:")
for i in inputs:
    print(f"Input: {i} -> Prediction: {nn.predict(i)}")
```
